# **Webscraping Conteúdo das rampas**





*   A abordagem para coletar o conteúdo do Guia 4 ventos foi pelo site map. https://guia4ventos.com.br/sitemap_index.xml
* Então o primeiro passo foi # salvar numa lista todas as urls nas páginas https://guia4ventos.com.br/post-sitemap1.xml











```python
# salvar numa lista todas as urls nas páginas https://guia4ventos.com.br/post-sitemap1.xml e https://guia4ventos.com.br/post-sitemap2.xml



from bs4 import BeautifulSoup
import requests

lista_rampas = []

pagina = requests.get("https://guia4ventos.com.br/post-sitemap1.xml")
soup = BeautifulSoup(pagina.text, "html.parser")

# Encontre todas as URLs nos elementos <loc> dentro do sitemap XML

urls = soup.find_all("loc")

for url in urls:
    lista_rampas.append(url.get_text())

print(lista_rampas)
len(lista_rampas)
```

- O segundo passo foi fazer a mesma coisa para a segunda url

```python
# salvar numa lista todas as urls nas páginas https://guia4ventos.com.br/post-sitemap1.xml e https://guia4ventos.com.br/post-sitemap2.xml


lista_rampas2 = []

pagina2 = requests.get("https://guia4ventos.com.br/post-sitemap2.xml ")
soup2 = BeautifulSoup(pagina2.text, "html.parser")

# Encontre todas as URLs nos elementos <loc> dentro do sitemap XML

urls2 = soup2.find_all("loc")

for url in urls2:
    lista_rampas2.append(url.get_text())

print(lista_rampas2)

```

- O terceiro passo foi concatenar as duas listas

```python
# concatenar as duas listas

lista_urls = lista_rampas + lista_rampas2
print(lista_urls)
len(lista_urls)
```

- O quarto passo foi pegar os titulos e conteudo dos artigos de cada url e salvar na lista

```python
# pegar o título e conteúdo de cada url da lista de urls

from bs4 import BeautifulSoup
import requests


conteudo = []

for url in lista_urls:
    # Acesse a URL atual
    response = requests.get(url)

    # Verifique se a solicitação foi bem-sucedida
    if response.status_code == 200:
        site = BeautifulSoup(response.text, "html.parser")

        # Obtenha o título da página
        titulo = site.find("h1", class_="entry-title").get_text()

        # Obtenha o conteúdo do corpo da página
        corpo = site.find_all("div", class_="et_pb_row et_pb_row_0")

        # Adicione o título e o conteúdo à lista "conteudo"
        conteudo.append(titulo)
        conteudo.extend([c.get_text() for c in corpo])

# Após o loop, a lista "conteudo" conterá os títulos e conteúdos de todas as páginas
print(conteudo)



```
